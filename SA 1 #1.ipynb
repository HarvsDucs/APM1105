```python
import requests

# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')
```

```python
with open('words.txt', 'r') as f:
    longest_word = ''
    for word in f:
        word = word.strip()
        if len(word) > len(longest_word):
            longest_word = word
print(f'The longest word is: {longest_word}')
```

```python
with open('words.txt', 'r') as f:
    letter = input('Enter a letter: ')
    count = 0
    for word in f:
        word = word.strip()
        if word[0] == letter:
            count += 1
print(f'The number of words starting with {letter} is: {count}')
```

```python
with open('words.txt', 'r') as f:
    substring = input('Enter a substring: ')
    for word in f:
        word = word.strip()
        # Check if the substring is present in the word
        if substring in word:
            print(word)
```

```python
with open('words.txt', 'r') as f:
    for word in f:
        word = word.strip()
        if word == word[::-1]:
            print(word)
```

```python
letter_count = {}
with open('words.txt', 'r') as f:
    # Loop over each line in the file
    for word in f:
        word = word.strip()
        for letter in word:
            if letter not in letter_count:
                letter_count[letter] = 1
            else:
                letter_count[letter] += 1
for letter, count in letter_count.items():
    print(letter, count)
```

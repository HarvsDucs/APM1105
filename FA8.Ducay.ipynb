```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew, sem
df = pd.read_csv("https://raw.githubusercontent.com/Apress/data-analysis-and-visualization-using-python/master/Ch07/Salaries.csv")
df.head(10)
profs_salary = df.groupby('rank')[['salary']]
```

```python
profs_salary.describe()
#mean and std is seen below
```

```python
valid_values = profs_salary.count()
valid_values
```

```python
missing_values = profs_salary.agg(lambda x: x.isna().sum())
missing_values
```

```python
profs_salary.median()
```

```python
profs_salary.mean()
#mean and std is seen below
```

```python
profs_salary.std()
```

```python
std_error_mean = profs_salary.std()/(profs_salary.count())**(1/2)
std_error_mean
```

```python
coeff_variation = profs_salary.std()/profs_salary.mean()
coeff_variation
```

```python
profs_salary.skew()
```

```python
n = profs_salary.count()
standard_error_skewness = np.sqrt((6 * n * (n-1)) / ((n-2) * (n+1) * (n+3)))
standard_error_skewness
```

```python
kurtosis = df.groupby('rank')[['salary']].agg(kurtosis)
kurtosis
#Note that the values were kinda different from values obtained from JASP which is -1.614		-1.011		0.201
```

```python
n = profs_salary.count()
std_error_kurtosis = ((24 * n * (n-2) * (n-3)) / ((n+3) * (n+5) * (n-3) * (n-4)))**(1/2)
std_error_kurtosis
#Note that the values were kinda different from values obtained from JASP which is 1.191		1.014		0.688
```

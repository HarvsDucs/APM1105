```python
import time

def iterative_exponent(base, exponent):
    final_answer = 1
    if base == 0 and exponent <= 0:
        return "Undefined"
    elif base == 0 and exponent > 0:
        return 0
    elif base > 0 or base < 0:
        if exponent == 0:
            if base < 0:
                return -1
            elif base > 0:
                return 1
        elif exponent > 0:
            while exponent != 0:
                final_answer *= base
                exponent -= 1
            return final_answer
        elif exponent < 0:
            while exponent != 0:
                final_answer *= 1/base
                exponent += 1
            return final_answer
    else:
        return "Please enter valid input"
    
base = int(input("Enter the base digit: "))
exponent = int(input("Enter the exponent digit: "))

start_time = time.time()
iterative_exponent(base, exponent)  
end_time = time.time()
total_time = end_time - start_time

print(iterative_exponent(base, exponent))
print(f"Iteration time total is {total_time}")
```

```python
import time

def recursive_exponent(base, exponent):
    final_answer = 1
    if base == 0 and exponent <= 0:
        return "Undefined"
    elif exponent == 0:
        if base < 0:
            return -1
        elif base > 0:
            return 1
    elif base == 0 and exponent > 0:
        return 0
    elif base > 0 or base < 0:
        if exponent > 0:
            return base * recursive_exponent(base, exponent-1)
        elif exponent < 0:
            return (1/base) * recursive_exponent(base, exponent+1)
    else:
        return "Please enter valid input"
    
base = int(input("Enter the base digit: "))
exponent = int(input("Enter the exponent digit: "))

start_time = time.time()
recursive_exponent(base, exponent)  
end_time = time.time()
total_time = end_time - start_time

print(recursive_exponent(base, exponent))
print(f"Recursion time total is {total_time}")
```

```python
import time

def sum_powers_iterative(n):
    sum = 0
    for i in range(n):
        sum += 2**i
    return sum

n = int(input("Enter the exponent digit n: "))

start_time = time.time()
iterative_result = sum_powers_iterative(n)
end_time = time.time()

iterative_time = end_time - start_time
print(f"Iterative result: {iterative_result}, time: {iterative_time}")
```

```python
import time

def sum_powers_recursive(n):
    if n == 0:
        return 1
    elif n < 0:
        return "Enter a non negative exponent"
    elif n > 0:
        return 2**(n-1) + sum_powers_recursive(n-1)

n = int(input("Enter the exponent digit n: "))

start_time = time.time()
recursive_result = sum_powers_recursive(n)
end_time = time.time()
recursive_time = end_time - start_time

print(f"Recursive result: {recursive_result}, time: {recursive_time}")
```

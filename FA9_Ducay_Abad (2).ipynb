```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_csv("C:/Users/Harvey/Downloads/Spotify_Youtube.csv")
display(df.head(10))
print(f"The columns in the dataset are: \n\n {df.columns}.")
print(f"\n\nThere are {df.shape[0]} datapoints from the dataset.")

columns_null = df.columns[df.isnull().any()]
print(f"Columns with null values are: \n\n{columns_null}")
columns_null = df.columns[df.isna().any()]
print(f"Columns with null values are: \n\n{columns_null}")

print(df.shape)

types_data = df.dtypes.value_counts()
print(f"\nData type counts are: \n\n{types_data}")
```

### Cleaning the data through managing missing values

```python
print(f"Total number of NA values in likes column before change: {df['Likes'].isna().sum()}")
df['Likes'].fillna(0, inplace=True)
print(f"Total number of NA values in likes column after change: {df['Likes'].isna().sum()}")

print(f"Total number of NA values in comments column before change: {df['Comments'].isna().sum()}")
df['Comments'].fillna(0, inplace=True)
print(f"Total number of NA values in comments column after change: {df['Comments'].isna().sum()}")

null_values = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
       'Duration_ms', 'Url_youtube', 'Title', 'Channel', 'Views', 'Description', 'Licensed', 'official_video', 'Stream']

for column in null_values:
    print(f"\n\nTotal number of NA values in {column} column before change: {df[column].isna().sum()}")
    df.dropna(subset=[column], inplace=True)
    print(f"\nTotal number of NA values in {column} columnd after change: {df[column].isna().sum()}")
```

### Exploring the album types column

```python
display(df.describe())
album_values = df['Album_type'].unique()
print(f"The set of values for Album Types are {album_values}")
album_count = df['Album_type'].value_counts()
print(f"\nThe number of count per value in Album Type column is: \n\n{album_count}")
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(9, 9))
plt.pie(album_count.values, labels=album_count.index, autopct='%1.1f%%', startangle=50)
plt.title("Album Types", fontsize=15)
plt.legend(fontsize=11)
# Show the chart
plt.show()
```

## Analyzing songs by artist and views-youtube, stream-spotify

```python
df.head(5)
```

```python
spotify_data = df.groupby('Track')[['Stream']].sum()
youtube_data = df.groupby('Track')[['Views']].sum()

artist_views = df.groupby('Artist')['Views'].sum()
artist_views_million = artist_views/1000000

artist_views_million = pd.DataFrame(artist_views_million)
artist_views_million_sorted = artist_views_million.sort_values('Views', ascending=False)

artist_streams = df.groupby('Artist')['Stream'].sum()
artist_streams_million = artist_streams/1000000

artist_streams_million = pd.DataFrame(artist_streams_million)
artist_streams_million_sorted = artist_streams_million.sort_values('Stream', ascending=False)

top_10_youtube = artist_views_million_sorted.iloc[0:10]
top_10_spotify = artist_streams_million_sorted.iloc[0:10]

print(spotify_data)
print(youtube_data)

display(top_10_youtube)
display(top_10_spotify)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,6))

sns.barplot(data=top_10_youtube, x=top_10_youtube.index, y='Views', ax=ax1)
ax1.set_title('Top 10 most Artists on YouTube', fontsize=26)
ax1.set_xlabel('Artist', fontsize=25)
ax1.set_ylabel('Views (in millions)', fontsize=25)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)

# second subplot
sns.barplot(data=top_10_spotify, x=top_10_spotify.index, y='Stream', ax=ax2)
ax2.set_title('Top 10 most streamed Artists on Spotify', fontsize=26)
ax2.set_xlabel('Artist', fontsize=25)
ax2.set_ylabel('Streams (in millions)', fontsize=25)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)

plt.show()
```

Only Ed Sheeran and Coldplay managed to be top viewed and streamed at the same time. 
The gap between the most viewed and most streamed artists isn't that big according to visualization

### Analyzing Top 10 songs on the basis of views, comments, and likes: with titles Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes

```python
df.head(3)
```

```python
most_viewed = df.groupby('Track')[['Views']].sum()
most_commented = df.groupby('Track')[['Comments']].sum()
most_liked = df.groupby('Track')[['Likes']].sum()

most_viewed_sorted = most_viewed.sort_values('Views', ascending=False)
most_commented_sorted = most_commented.sort_values('Comments', ascending=False)
most_liked_sorted = most_liked.sort_values('Likes', ascending=False)

most_viewed_sorted_10 = most_viewed_sorted.iloc[0:10]
most_commented_sorted_10 = most_commented_sorted.iloc[0:10]
most_liked_sorted_10 = most_liked_sorted.iloc[0:10]

most_viewed_sorted_10["Views_millions"] = most_viewed_sorted_10["Views"]/1000000
most_commented_sorted_10["Comments_millions"] = most_commented_sorted_10["Comments"]/1000000
most_liked_sorted_10["Likes_millions"] = most_liked_sorted_10["Likes"]/1000000

display(most_viewed_sorted_10)
display(most_commented_sorted_10)
display(most_liked_sorted_10)
```

### Plotting most viewed track graph

```python
plt.figure(figsize=(20,7))
sns.barplot(x=most_viewed_sorted_10.index, y="Views_millions", data = most_viewed_sorted_10)

plt.margins(x=0.01)
plt.title("Top 10 Tracks by Views", fontsize=30)
plt.xlabel("Tracks", fontsize=25)            
plt.ylabel("Number of Views in Millions", fontsize=25)
plt.xticks(rotation=90, fontsize= 15)
```

As in the amount of likes so as the amount of views, Despacito and See You Again leads far off from the other tracks that were part of the top 10 most viewed.

## Exploring single releases by artists and identify top performers based on likes and display a table for each artist with their Views and Likes as the other columns

```python
singles_data = df.loc[df['Album_type'] == 'single']

artist_data = singles_data.groupby('Artist')[['Likes', 'Views']].sum()

artist_data_likes = artist_data.sort_values('Likes',ascending=False)
artist_data_views = artist_data.sort_values('Views',ascending=False)

artist_data_likes_10 = artist_data_likes.iloc[0:10]
artist_data_views_10 = artist_data_views.iloc[0:10]

artist_data_likes_10["Likes_million"] = artist_data_likes_10["Likes"]/1000000
artist_data_views_10["Views_million"] = artist_data_views_10["Views"]/1000000

# Display the merged data
display(artist_data_likes_10)
display(artist_data_views_10)
```

### Remarks about Views, Likes, and the Artist
As we can see, majority of those in the top 10 most liked artists are also included in the top 10 most viewed artists. It can be intuitively said from the data that these two are positively correlated. The top 3 most viewed artists with singles are also the top 3 artists with singles that are liked.

## Exploring compilation releases by artists and identifying top performers based on views and likes and displaying a table for each artist with their Views and Likes as the other columns.

```python
compilation_data = df.loc[df['Album_type'] == 'compilation']

artist_data = compilation_data.groupby('Artist')[['Likes', 'Views']].sum()

artist_data_likes = artist_data.sort_values('Likes',ascending=False)
artist_data_views = artist_data.sort_values('Views',ascending=False)

artist_data_likes_10 = artist_data_likes.iloc[0:10]
artist_data_views_10 = artist_data_views.iloc[0:10]

artist_data_likes_10["Likes_million"] = artist_data_likes_10["Likes"]/1000000
artist_data_views_10["Views_million"] = artist_data_views_10["Views"]/1000000

# Display the merged data
display(artist_data_likes_10)
display(artist_data_views_10)
```

### Remarks about Views, Likes, and the Artist
As we can see, majority of those in the top 10 most liked artists are also included in the top 10 most viewed artists. It can be intuitively said from the data that these two are positively correlated. The top 3 most viewed artists with compilation are also the top 3 artists with compilation that are liked. Singles and compilation data are both positively correlated with its likes and views, although top artists with singles differ from top artists with compilation.
